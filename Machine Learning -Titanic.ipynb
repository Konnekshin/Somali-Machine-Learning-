### Tilaabada Sadexaad (Step 4)

##### Xog Nadiifi oo kala habee (Clean Data)


```python
# Xogta loo isticmaalayo tabo baridda (Training data)
df_train= train.dropna(axis=0, subset=['Embarked', 'Age'])
df_train
df_train.isnull().sum()
```

```python
df_train['Title'] = df_train['Name'].str.split(',', expand=True)[1].str.split('.', expand=True)[0]
df_train.head(50)
```

```python
df_train= df_train.drop(['Name', 'Cabin', 'Ticket', 'PassengerId'] ,axis=1)
df_train.head()
```

```python
x_train = df_train.drop("Survived",axis=1)
y_train = df_train[['Survived']]

x_train.head()
y_train.head()
```

```python
test_df=test
```

```python
test['Age'] = test['Age'].interpolate()
test['Fare'] = test['Fare'].interpolate()
test.isnull().sum()
```

```python
# Xogta loo isticmaalayo tijaabinta (Testing data)
#df_test = test.dropna(axis=0, subset=['Fare', 'Age'])
df_test=test
```

```python
df_test['Title'] = df_test['Name'].str.split(',', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df_test= df_test.drop(['Name', 'Cabin', 'Ticket', 'PassengerId'] ,axis=1)
```

```python
df_test.head()
df_test.shape
```

### Tilaabada Sadexaad (Step 5)

##### U badal wixii xaruuf ah tiro   (Convert leters to numeric)

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
# Xogta tabobaridda loo badaley tiro (Training data letters to numeric)
x_train.iloc[:,1]= le.fit_transform(x_train.iloc[:,1].values)

x_train.iloc[:,6]= le.fit_transform(x_train.iloc[:,6].values)

x_train.iloc[:,7]= le.fit_transform(x_train.iloc[:,7].values)

x_train.head()

#plt.figure(figsize=(14,12))
# sns.heatmap(x_train.astype(float).corr(),linewidths=0.1, square=True,  linecolor='white', annot=True)
```

```python
# Xogta tijaabinta loo badaley tiro (Testing data letter to numeric)
df_test.iloc[:,1]= le.fit_transform(df_test.iloc[:,1].values)

df_test.iloc[:,6]= le.fit_transform(df_test.iloc[:,6].values)

df_test.iloc[:,7]= le.fit_transform(df_test.iloc[:,7].values)
```

```python
df_test.head()
```

### Tilaabada Sadexaad (Step 6)

##### Dhisidda modelka / Xisaabta (Modelling / Using Algorithm)

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
# model training
lr=lr.fit(x_train, y_train)

# Accuracy Check
lr_score = lr.score(x_train, y_train)

# Prediction testing data
lr_predict=lr.predict(df_test)
lr_score
```

```python
lr_predict
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,
                            criterion = "gini",
                            max_features='auto')
```

```python
rf= rf.fit(x_train, y_train)

rf_predict=rf.predict(df_test)

rf_score= rf.score(x_train, y_train)

rf_score
```

```python
# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)

```

```python
knn.fit(x_train, y_train)

knn_score = knn.score(x_train, y_train)

knn_predict = knn.predict(df_test)

```

```python
knn_score
```

### Tilaabada Sadexaad (Step 7)

##### Iska hubi modelka shaqadiisa (Evaluating the Models)

```python
# Cross Validation
from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf, x_train, y_train, cv=10, scoring = "accuracy")

scores
scores.mean()

```

```python
# Confusion Matrix



from sklearn.model_selection import cross_val_predict

from sklearn.metrics import confusion_matrix
predictions = cross_val_predict(rf, x_train, y_train, cv=3)
confusion_matrix(y_train, predictions)
```

```python
# TN, FP, FN & TP
```

```python
# Precision
# Recall

from sklearn.metrics import precision_score, recall_score

print("Precision:", precision_score(y_train, predictions))
print("Recall:",recall_score(y_train, predictions))
```

```python
# F-Score

from sklearn.metrics import f1_score
f1_score(y_train, predictions)
```
